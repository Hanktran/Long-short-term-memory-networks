```python
from math import sin
from math import pi
from math import exp
from matplotlib import pyplot
```

```python
# create sequence
length = 100
period = 10
decay = 0.05
sequence = [0.5 + 0.5 * sin(2 * pi * i / period) * exp(-decay * i) for i in range(length)]
```

```python
# plot sequence
pyplot.plot(sequence)
pyplot.show()
```
